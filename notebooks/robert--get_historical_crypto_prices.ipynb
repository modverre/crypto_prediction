# coingecko

```python
api_url = 'https://api.coingecko.com/api/v3'
api_endpoint = '/coins/list'
```

```python
request = requests.get(f'{api_url}{api_endpoint}')
```

```python
coins = request.json()
```

```python
len(coins)
```

```python
for coin in coins:
    if coin['symbol'] == 'doge':
        print(coin)
```
